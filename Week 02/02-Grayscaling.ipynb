## <div style="direction:rtl;text-align:right;font-family:B Lotus, B Nazanin, Tahoma">Converting images to grayscale or Grayscaling </div>

#### Grayscaling is process by which an image is converted from a full color to shades of grey (black & white)

In OpenCV, many functions grayscale images before processing. This is done because it simplifies the image, acting almost as a noise reduction and increasing processing time as there is less information in the image.

### Let convert our color image to greyscale

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import cv2

# Load our input image
image = cv2.imread('/content/drive/MyDrive/DIP/New_Zealand_Boat.jpg')
cv2.imshow('Original', image)
cv2.waitKey()

# We use cvtColor, to convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow('Grayscale', gray_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Another method faster method
img = cv2.imread('/content/drive/MyDrive/DIP/New_Zealand_Boat.jpg',0)
cv2.imshow('Grayscale', img)
cv2.waitKey()
cv2.destroyAllWindows()

print(img.shape)
```

Now, let's take a look at the dimensions of the color and grayscale images!</div>

```python
image = cv2.imread('/content/drive/MyDrive/DIP/New_Zealand_Boat.jpg')

#Let's look at the individual color levels for the first pixel (10,50)
B, G, R = image[10, 50]
print (B, G, R)
print ("shape:", image.shape)
```
